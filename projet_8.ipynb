### A - Importation des librairies :

```python
# basics
import os
from os import listdir
import pandas as pd
import matplotlib.pyplot as plt
```

### B - Importation et transformation des fichiers :

#### 1) Adresse des répertoires :

```python
# Adresse des répertoires :
path = "C:/Users/7700k/Desktop/Projet 8/dataset/"
cat = listdir(path)
```

#### 2) Création d'un Dataframe :

```python
# Création d'un DataFrame
data = []
for c in cat :
    for file in listdir(path + c) :
        data.append(c + "/" + file)
data = pd.DataFrame(data)
data.rename(columns={0:"images"}, inplace=True)
data["target"] = data["images"].str.split("/").str[0]
```

#### 3) Aperçu :

```python
def list_fct(name) :
    list_image_name = [data["images"][i] for i in range(len(data)) if data["target"][i]==name]
    return list_image_name

for name in data["target"].unique() :
    print(name)
    # print("-------")
    for i in range(3):
        plt.subplot(130 + 1 + i)
        filename = path + "/" + list_fct(name)[i+10]
        image = plt.imread(filename)
        plt.imshow(image)
        plt.axis("off")
    plt.show()
```
